```python
import os
import tifffile as tiff

input_folder = 'data'
brightfield_folder = 'brightfield'
fluo_folder = 'fluo'

# Create output folders if they don't exist
os.makedirs(brightfield_folder, exist_ok=True)
os.makedirs(fluo_folder, exist_ok=True)

for filename in os.listdir(input_folder):
    if filename.lower().endswith('.tif') or filename.lower().endswith('.tiff'):
        filepath = os.path.join(input_folder, filename)
        image = tiff.imread(filepath)

        # Make sure the image has at least 2 channels
        if image.ndim >= 3 and image.shape[0] >= 2:
            brightfield = image[0]
            fluo = image[1]

            tiff.imwrite(os.path.join(brightfield_folder, filename), brightfield)
            tiff.imwrite(os.path.join(fluo_folder, filename), fluo)
        else:
            print(f"Skipping {filename}: not enough channels")
```
